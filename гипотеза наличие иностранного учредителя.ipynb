```python
import requests

req = requests.get('https://egrul.itsoft.ru/'+inn+'.json')
```

```python
import json

req_struct = json.loads(req.text)
```

```python
# Код статуса:  0 - не определено     
#               1 - Юр лица
#               2 - Физ лица
#               3 - Держатель реестра
#               4 - Явно иностраный капитал
stat_code = 0 
```

```python
if req_struct['СвЮЛ'].get('СвДержРеестрАО'):
    stat_code = 3
elif req_struct['СвЮЛ'].get('СвУчредит'):
    if req_struct['СвЮЛ']['СвУчредит'].get('УчрЮЛИн'):
        if req_struct['СвЮЛ']['СвУчредит']['УчрЮЛИн'].get('СвРегИн'):
            stat_code = 4
        else:
            stat_code = 1
    else:
        stat_code = 2        
```
